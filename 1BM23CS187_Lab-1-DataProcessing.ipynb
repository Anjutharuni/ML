```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

```

```python
df = pd.read_csv("Diabetes.csv")

print("Diabetes Dataset Info:")
df.info()

```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].mean())

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna(df[col].mode()[0])

```

```python
encoder = LabelEncoder()
cat_cols = ["Gender", "CLASS"]

for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])

print("Categorical Data Encoded")
```

```python
def remove_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[col] >= lower) & (data[col] <= upper)]

for col in df.select_dtypes(include=np.number):
    df = remove_outliers(df, col)

print("Outliers Removed")
```

```python
minmax = MinMaxScaler()

df_minmax = df.copy()

features = df_minmax.drop("CLASS", axis=1)

df_minmax[features.columns] = minmax.fit_transform(features)

print("Min-Max Normalized Data:")
df_minmax.head()
```

```python
standard = StandardScaler()

df_standard = df.copy()

features = df_standard.drop("CLASS", axis=1)

df_standard[features.columns] = standard.fit_transform(features)

print("Standard Scaled Data:")
df_standard.head()
```

```python
df2 = pd.read_csv("adult.csv")

print("Dataset Loaded Successfully")
df2.head()
```

```python
df2 = df2.replace(" ?", np.nan)
df2 = df2.replace("?", np.nan)
print(df2.isnull().sum())
```

```python
# Fill numeric columns with mean
for col in df2.select_dtypes(include=np.number).columns:
    df2[col] = df2[col].fillna(df2[col].mean())

# Fill categorical columns with mode
for col in df2.select_dtypes(include="object").columns:
    df2[col] = df2[col].fillna(df2[col].mode()[0])
```

```python
encoder = LabelEncoder()

for col in df2.select_dtypes(include="object").columns:
    df2[col] = encoder.fit_transform(df2[col])

print("Categorical Columns Encoded")
```

```python
def remove_outliers(data, col):

    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[col] >= lower) & (data[col] <= upper)]

for col in df2.select_dtypes(include=np.number).columns:
    df2 = remove_outliers(df2, col)

print("Outliers Removed")
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

df_minmax = df2.copy()

X = df_minmax.iloc[:, :-1].astype(float)

df_minmax.iloc[:, :-1] = minmax.fit_transform(X)

print("Min-Max Scaled Data:")
df_minmax.head()

```

```python
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()

df_standard = df2.copy()

X = df_standard.iloc[:, :-1].astype(float)

df_standard.iloc[:, :-1] = standard.fit_transform(X)

print("Standard Scaled Data:")
df_standard.head()
```
